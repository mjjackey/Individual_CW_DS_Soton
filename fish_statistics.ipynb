```r
lm1<- lm(V2~V1)
summary(lm1)
```

```r
lm2<- lm(V2~poly(V1,2))
summary(lm2)
```

```r
lm3<- lm(V2~poly(V1,3))
summary(lm3)
```

```r
lm0 <- lm (V2 ~ 1)
AIC(lm0,lm1,lm2,lm3)
```

```r
attach(fish.data)
# plot(x=V1,y=V2)  
# lines (predict(lm1) ~ V1, col="red")  
# lines (predict(lm2) ~ V1, col="blue")
df <- data.frame (x=sort(V1), y=V2[order(V1)])
# df
lm1 <- lm ( df$y ~ df$x)
lm2 <- lm ( df$y ~ poly (df$x, 2))
pdf(file = "fitting-curve.pdf")
plot (df$y ~ df$x, main="Scatterplot Sizes of Catch vs Time of Catch and Fitting Curve", xlab="Time of Catch (h)", ylab="Sizes of Catch (kg)")
lines (predict (lm2) ~ df$x, col="blue")
lines (predict (lm1) ~ df$x, col="red")
dev.off()
```

### 4.2 分析V3和V2的关系，V3对V1和V2的影响

```r
flm0 <- lm (V2 ~ 1)
flm1 <- lm (V2 ~ V3)
summary (flm1)
AIC(flm0,flm1)
```

```r
# input <- fish.data[,c("V1","V2","V3")]
result <- aov(V2~V1*V3,data = fish.data)
print(summary(result))
```

```r
result <- aov(V2~V1+V3,data = fish.data)
print(summary(result))
```

```r
result1 <- aov(V2~V1*V3,data = fish.data)
result2 <- aov(V2~V1+V3,data = fish.data)
print(anova(result1,result2))
```
